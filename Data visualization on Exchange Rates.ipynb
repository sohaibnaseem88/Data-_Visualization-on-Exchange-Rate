Storytelling Data Visualization on Exchange Rate

# It enable the identification of long-term trends in Euro to USD exchange rates, whether it's a sustained appreciation or depreciation of one currency against the other.





```python
import pandas as pd
import warnings

warnings.filterwarnings("ignore")
```

```python
exchange_rates = pd.read_csv("euro-daily-hist_1999_2022.csv")
exchange_rates.shape
```

```python
exchange_rates.head()
```

```python
exchange_rates.info()
```

```python
exchange_rates.rename(columns={'[US dollar ]' : 'US_dollar ','Period\\Unit:': 'Time'}, inplace=True)
```

```python
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
```

```python
exchange_rates.sort_values('Time', inplace=True)
```

```python
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar ']].copy()
euro_to_dollar['US_dollar '].value_counts()   # 62 '-' characters
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar '] != '-']
euro_to_dollar['US_dollar '] = euro_to_dollar['US_dollar '].astype(float)
euro_to_dollar.info()
```

```python
# Rolling Mean
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar '])
plt.show()
```

```python
values = pd.DataFrame()
values['daily_values'] = pd.Series(range(1,20,2))
```

```python
values
```

```python
values['rolling_mean_2'] = values['daily_values'].rolling(2).mean()
values
```

```python
values['rolling_mean_3'] = values['daily_values'].rolling(3).mean()
values['rolling_mean_5'] = values['daily_values'].rolling(5).mean()
values
```

```python
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar '])
plt.title('Original values', weight='bold')

for i, rolling_mean in zip([2, 3, 4, 5, 6],
                           [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
               euro_to_dollar['US_dollar '].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')
    
plt.tight_layout() # Auto.adjusts the podding between subplots
plt.show()
```

```python
# Coming up with an idea 
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar '].rolling(30).mean()
euro_to_dollar
```

```python
# Storytelling Data  Visualization
# Financial Crisis 2007-2008
```

```python
financial_crisis = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2006
                   ) & (euro_to_dollar['Time'].dt.year <= 2009)]
financial_crisis_7_8 = euro_to_dollar.copy(
                   )[(euro_to_dollar.Time.dt.year >=2007
                   ) & (euro_to_dollar.Time.dt.year <=2008)]
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')

### Adding the plot
fig,ax = plt.subplots(figsize=(8,3))
ax.plot(financial_crisis['Time'],
        financial_crisis['rolling_mean'],
        linewidth=1, color='#A60785')

### Highlighting the 2007-2008 period
ax.plot(financial_crisis_7_8['Time'],
       financial_crisis_7_8['rolling_mean'],
       linewidth=3, color='#e23d28')

# print (ax.set_xticklabels(["2006", "", "2007", "", "2008", "", "2009", "", "2010"], alpha = 0.3))
 
ax.set_xticklabels([])

x = 0.02
for year in ['2006','2007' ,'2008', '2009' , '2010']:
    ax.text(x, -0.08, year, alpha=0.5, fontsize=11, transform = plt.gca().transAxes)
    x += 0.22888
             
# print(plt,yticks())
             
ax.set_yticklabels([])             

y =  0.07
for rate in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(-.04, y, rate, alpha=0.5, fontsize=11, transform = plt.gca(). transAxes)
    y += 0.2333
    
    
### Adding a title  and a subtitle
ax.text(-.05, 1.2, "Euro.USD rate peaked at 1.59 during 2007-2008's finanical crisis",
       weight="bold", transform = plt.gca().transAxes)
ax.text(-.05, 1.1, "Euro.USD exchange rates between 2006 and 2010",
        size=12, transform = plt.gca().transAxes)


## Adding a signature
ax.text(-.05,-0.25, 'SOHAIB NASEEM' + ' '*80 + 'Source: European Central Bank', 
        color = '#f0f0f0', backgroundcolor = '#404d4d',
        size=12, transform = plt.gca().transAxes)

ax.axvspan(xmin=pd.to_datetime("2008-04-01"), xmax=pd.to_datetime("2008-09-01"), ymin=0.05,
          alpha=0.3, color='grey')

plt.show()
```

```python
### Euro to USD Rate During Different Govt. Era
```

```python
all_3 = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2001) & (euro_to_dollar['Time'].dt.year < 2021)]
bush = all_3.copy()[(all_3['Time'].dt.year < 2009)]
obama = all_3.copy()[all_3['Time'].dt.year >= 2009 & (all_3['Time'].dt.year < 2017)]
trump = all_3.copy()[all_3['Time'].dt.year >= 2017 & (all_3['Time'].dt.year < 2021)]
```

```python
style.use('fivethirtyeight')

plt.figure(figsize=(14,8))

# for subplots
ax1 = plt.subplot(3,3,1)
ax2 = plt.subplot(3,3,2)
ax3 = plt.subplot(3,3,3)

#for large plot
ax4 = plt.subplot(3,1,2)

axes = [ax1, ax2, ax3, ax4]
for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2', '1.4', '1.6 $'],alpha=0.5)

#Bush
ax1.plot(bush['Time'],bush['rolling_mean'],linewidth=2,color='#c6eb34')
ax1.set_xticklabels(['','2001','','2003','','2005','','2007','','2009'],fontsize=12)
ax1.text(0.11,2.4,'Bush',fontsize=20,color='salmon',transform=plt.gca().transAxes)
ax1.text(0.11,2.2,'2001-2009',fontsize=18,color='salmon',transform=plt.gca().transAxes)

#Obama
ax2.plot(obama['Time'],obama['rolling_mean'],linewidth=2,color='#34bdeb')
ax2.set_xticklabels(['','2009','','2011','','2013','','2015','','2017'],fontsize=12)
ax2.text(0.43,2.4,'Obama',fontsize=20,color='salmon',transform=plt.gca().transAxes)
ax2.text(0.43,2.2,'2009-2017',fontsize=18,color='salmon',transform=plt.gca().transAxes)

#Trump
ax3.plot(trump['Time'],trump['rolling_mean'],linewidth=2,color='#5634eb')
ax3.set_xticklabels(['','2017','','2018','','2019','','2020','','2021'],fontsize=12)
ax3.text(0.83,2.4,'Trump',fontsize=20,color='salmon',transform=plt.gca().transAxes)
ax3.text(0.83,2.2,'2017-2021',fontsize=18,color='salmon',transform=plt.gca().transAxes)

#merging
ax4.plot(bush['Time'],bush['rolling_mean'])
ax4.plot(obama['Time'],obama['rolling_mean'])
ax4.plot(trump['Time'],trump['rolling_mean'])

ax1.text(0.2, 2.8, ' Euro to USD Rates During Three Presidents Era', fontsize=20, weight='bold', transform=plt.gca().transAxes)
ax4.text(0.02, -0.35,'Sohaib Naseem' + ' '*133 + 'Techma Zone', size=14,weight='bold',color='white',backgroundcolor='grey',alpha=0.5, transform=plt.gca().transAxes)
ax.axvspan(xmin=pd.to_datetime("2008-04-1"), xmax=pd.to_datetime("2008-09-1"),ymin=0.09, color='#f5e74e',alpha=0.5)
plt.show()
```
